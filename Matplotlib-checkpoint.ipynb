```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x= np.linspace(4,15,100)
y= np.sin(x)  #its value of sin theta
```

#Lineplot

```python
plt.figure(figsize=(5,3))
plt.plot(x,y, label="Sin Wave",linestyle="-",linewidth=2,color="#234567")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.title("Sin Wave")
plt.legend()
plt.grid(True)
```

Barchart plot


```python
category = ["A","B","C","D","E"]
value = [33,54,42,51,29]
```

```python
plt.figure(figsize=(5,3))
plt.bar(category,value,alpha=0.9,edgecolor="black",
        color=["#101010","#245300","#425025","#252525","#303020"])
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Comparing Categories")
plt.grid(True)
```

#Scatter Plot 
c : sets color
s: Specifies sixe of marker
colors -- viridus, plasma ,Blues 

```python
x= np.random.rand(100)
y= np.random.rand(100)
color= np.random.rand(100)
size = 1000 * np.random.rand(100)
```

```python
plt.scatter(x,y,c=color,s=size,alpha=0.7,cmap="plasma")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.title("Scatter Plot")
plt.colorbar(label="Color Intensity")
plt.grid(True)
```

### Histogram
### Histogram display the distribution of single numeric variable.
### bins : the number of bin in histogram
### Color  :
### edgecolor :
### alpha :

```python
dice = np.array([1,2,3,4,5,6]) #Dice
sample = np.random.choice(dice,4) #Rolling dice 4 times in a row and getting values
print("Output of 4 samples",sample)
print("Mean of above samples",sample.mean())
```

```python
samples_1000=[np.random.choice(dice,4).mean()for i in range(1000)]
samples_1000
```

```python
plt.figure(figsize=(5,3))
plt.hist(samples_1000, alpha =0.7, edgecolor="black",bins=100, histtype="barstacked")
```

```python
data = np.random.random(10000)
plt.hist(data, alpha =0.7, edgecolor="black",bins=100 , histtype="stepfilled")
plt.show()
```

```python
plt.hist(data, alpha =0.7, edgecolor="black",bins=100 , histtype="stepfilled")
plt.xlabel(" Values")
plt.ylabel("Frequency")
plt.title("Histogram")
plt.grid(True)
```

### Pie Chart
### it represents parts of a whole
### colors:Sets individual colors for segment
### autopct
### Startangle:
### labels:
###

```python
sizes = np.random.randint(35,89,4)
labels = ["East","West","North","South"]
plt.pie(sizes,labels=labels, colors=["gold","red","green","purple"],
       autopct="%1.1f%%", explode=[0,0,0.2,0])
plt.title("Piechart")
plt.show()
```

### Box plot

```python
data = np.random.normal(loc=0, scale=1,size=(100,))

plt.boxplot(data,patch_artist=True, vert=False)
plt.title("Box Plot")
plt.text
plt.show()
```

### Polar plot

```python
angles = np.linspace(0,2*np.pi,100)
radii = np.random.rand(100)
```

```python
ax =plt.subplot(111,polar=True)
ax.plot(angles,radii,color="lightblue",linewidth=0.1)
ax.fill(angles,radii,color= "blue",alpha=0.6)
plt.title("Polar Plot")
plt.show()
```
