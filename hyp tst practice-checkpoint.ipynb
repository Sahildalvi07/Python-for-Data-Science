```python
sns.histplot(Iris["sepal width (cm)"],kde=True,alpha=0.7)
plt.title("Iris Column Distribution")
plt.show()
```

```python
# H0 -> data is normally distributed
# Ha-> data is not normally distributed

#checking whether to accept the null hypotehsis or reject it.

sh_stat,sh_p = shapiro(Iris["sepal width (cm)"])
print("Shapiro Statistical Value :",sh_stat)
print("Shapiro p_value:",sh_p)

alpha = 0.05
if sh_p > alpha:
    print("data id noramally distributed (Acepting H0)")
else:
    print("data is not normally distributed (Rejecting H0)")
```
