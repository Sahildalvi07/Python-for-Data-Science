```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x= np.linspace(2,20,100)
y= np.sin(x)
```

```python
x
```

```python
y
```

### Lineplot

```python
plt.plot(x,y, label="Sine Wave",linestyle="--",linewidth=4,color="#707070")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.title("Sin Wave")
plt.legend()
plt.grid(True)
plt.show()
```

### Barchart

```python
Catergory =["A","B","C","D","E"]
Value = [25,35,45,55,65]
```

```python
plt.bar(Catergory,Value,alpha=0.9,edgecolor="black",
       color=["#102030","#256535","#707070","#303050","#252015"])
plt.xlabel("categories")
plt.ylabel("values")
plt.title("Catergories chart")
plt.grid(True)
```

### Scatter Plot

```python
x= np.random.rand(100)
y= np.random.rand(100)
color= np.random.rand(100)
size = 1000 * np.random.rand(100)
```

```python
plt.scatter(x,y,c=color,s=size,alpha=0.9,cmap="jet")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.title("Scatter Plot")
plt.colorbar(label="Color Intensity")
plt.grid(True)
```
