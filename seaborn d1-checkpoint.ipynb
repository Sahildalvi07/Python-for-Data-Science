```python
import seaborn as sns
import matplotlib.pyplot as plt
# from matplotib import pyplot as plt
```

```python
weight = [62,64,69,75,66,68,65,71,76,73]
height = [120,136,148,175,137,165,154,172,200,187]
```

```python
%matplotlib inline #also can done by notebook
plt.figure(figsize=(5,3)) # set size
plt.xlabel("height")
plt.ylabel("weight")
plt.title("height vs weight")
sns.scatterplot(x = height,y=weight)
plt.show()
```

```python
#countplot
gender = ["Female","Female","Female","Female",
         "Male", "Male", "Male", "Male", "Male", "Male"]
```

```python
sns.countplot(x=gender,hue=gender,legend=["Female", "Male"])
plt.title("Count of Male and Feamle")
plt.show()
```
